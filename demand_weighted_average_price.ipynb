## Demand weighted average price plot

Code to generate a suitable demand weighted average price plot (and comparison vs previous years) for reporting purposes.

```python
import pandas as pd
import datetime as dt
import EAtools as ea
ea.set_options()
ea.ea_report_style()
```

Connect to the Data Warehouse

```python
connection = ea.DW_connect(linux=True)
```

Define start and end datetime objects

```python
'''From the start of last month, to yesterday'''
dBegt0 = datetime.date(dt.datetime.now().date().year,dt.datetime.now().date().month-1,1)
dEndt0 = dt.datetime.now().date()-dt.timedelta(days=1)

'''From the start of last month, last year, to the end of the next month, last year...'''
dBegt1 = datetime.date(dt.datetime.now().date().year-1,dt.datetime.now().date().month-1,1)
dEndt1 = datetime.date(dt.datetime.now().date().year-1,dt.datetime.now().date().month+2,1)-dt.timedelta(days=1)

'''From the start of last month, 2 years ago, to the end of the next month 2 years ago...'''
dBegt2 = datetime.date(dt.datetime.now().date().year-2,dt.datetime.now().date().month-1,1)
dEndt2 = datetime.date(dt.datetime.now().date().year-2,dt.datetime.now().date().month+2,1)-dt.timedelta(days=1)
```

Get the data from the DW (*Warning...This is SLOW...*)

```python
l0 = ea.timeseries_convert(ea.FP_getter(connection,ea.query_demand(dBegt0,dEndt0)))
p0 = ea.timeseries_convert(ea.FP_getter(connection,ea.query_prices(dBegt0,dEndt0)))

l1 = ea.timeseries_convert(ea.FP_getter(connection,ea.query_demand(dBegt1,dEndt1)))
p1 = ea.timeseries_convert(ea.FP_getter(connection,ea.query_prices(dBegt1,dEndt1)))

l2 = ea.timeseries_convert(ea.FP_getter(connection,ea.query_demand(dBegt2,dEndt2)))
p2 = ea.timeseries_convert(ea.FP_getter(connection,ea.query_prices(dBegt2,dEndt2)))
```

```python
p0.head()
```

Calculate the LWAP

```python
def lwap(p,l):
    return (p*l).sum(axis=1)/l.sum(axis=1)

lwap0 = lwap(p0.price,l0.demand)
lwap1 = lwap(p1.price,l1.demand)
lwap2 = lwap(p2.price,l2.demand)
```

Shift the index of previous years data to match current year and whack into a DataFrame

```python
def shift_index(df,year_shift):
    return df.index.map(lambda x: dt.datetime(x.year+year_shift,x.month,x.day,x.hour,x.minute))

lwap1.index = shift_index(lwap1,1)
lwap2.index = shift_index(lwap2,2)
lwaps = pd.DataFrame({str(dBegt0) + ' to ' + str(dEndt0):lwap0,str(dBegt1) + ' to ' + str(dEndt1):lwap1,str(dBegt2) + ' to ' + str(dEndt2):lwap2})
```

Plot

```python
lwaps.plot(color=[ea.ea_p['gy1'],ea.ea_p['br2'],ea.ea_p['rd1']])
ylabel('$/MWh')
```

Examples of python text string manipulation...

```python
"The current demand weighted average price so far this month is $%.2f with a maximum of $%.2f on %s" % \
(lwap0.groupby(lambda x: x.month).mean()[0:1].values[0], \
 lwap0.groupby(lambda x: x.month).max()[0:1].values[0],\
 lwap0.groupby(lambda x: x.month).idxmax()[0:1].to_string().split()[1])
```
