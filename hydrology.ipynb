```python
link='''http://www.systemoperator.co.nz/f3933,83729137/HRC_data_2013_2014.xlsx'''
HRC_SO = pd.ExcelFile(urllib2.urlopen(link))
NZ_HRC_SO = HRC_SO.parse(HRC_SO.sheet_names[0], header=3).T.ix[:,:6].T
NZ_HRC_SO = NZ_HRC_SO.rename(index = dict(zip(NZ_HRC_SO.index,['1%','2%','4%','6%','8%','10%']))).T.applymap(lambda x: float(x))
SI_HRC_SO = HRC_SO.parse(HRC_SO.sheet_names[0], header=3).T.ix[:,9:15].T
SI_HRC_SO = SI_HRC_SO.rename(index = dict(zip(SI_HRC_SO.index,['1%','2%','4%','6%','8%','10%']))).T.applymap(lambda x: float(x))
```

#### Correct historic data held at the EA.  Note: the SO data is only accurate ex-anti (forward looking) and can't be currently relied upon for historic purposes...

```python
NZ_HRC = pd.read_csv('hydroriskcurves/HRC_History_20130618.csv',index_col=0,parse_dates=True,dayfirst=True,usecols=[0,1,2,3,4,5,6]).drop_duplicates()
SI_HRC = pd.read_csv('hydroriskcurves/HRC_History_20130618.csv',index_col=0,parse_dates=True,dayfirst=True,usecols=[0,8,9,10,11,12,13])
SI_HRC = SI_HRC.reset_index().rename(columns=dict(zip(SI_HRC.reset_index().columns,['South Island','1%','2%','4%','6%','8%','10%']))).set_index('South Island').drop_duplicates()
```

Use data from both sets - EA data before May, SO afterwards

```python
SI_HRC2 = SI_HRC_SO['6/2013':].append(SI_HRC['12/2012':'5/2013'])
NZ_HRC2 = NZ_HRC_SO['6/2013':].append(NZ_HRC['12/2012':'5/2013'])
```

```python
def hrc_plot(hrc,actual,means):
           
    fig = figure(1,figsize=[25,20])
    ax1 = fig.add_subplot(111)
    colours = {'1%':ea.ea_p['yl2'],'2%':ea.ea_p['yl1'],'4%':ea.ea_s['or2'],'6%':ea.ea_s['or1'],'8%':ea.ea_p['rd2'],'10%':ea.ea_p['rd1']}
    
    def hrc_plotter(HRC,actual,means,ax,ylabel,colours):

        x = HRC.index
        for c in HRC.columns:
            HRC[c].plot(color=colours[c])
            #ax.fill_between(x,HRC[c],color=colours[c])
            xlabels = ax.get_xticklabels() 
        HRC['10%'].plot(color=ea.ea_p['rd1'],linewidth=4)
        means.plot(color=ea.ea_s['gy2'],linewidth=6)
        actual.plot(color=ea.ea_p['bl1'],linewidth=4)
        for label in xlabels: 
            label.set_rotation(0) 
        ax.set_xlabel('')
        ax.set_ylabel(ylabel)
        act1 = Line2D((0,1),(0,0), color=ea.ea_p['bl1'],linewidth=4)
        mean1= Line2D((0,1),(0,0), color=ea.ea_p['gy2'],linewidth=6)
        hrc1 = Line2D((0,1),(0,0), color=colours['1%'])
        hrc2 = Line2D((0,1),(0,0), color=colours['2%'])
        hrc4 = Line2D((0,1),(0,0), color=colours['4%'])
        hrc6 = Line2D((0,1),(0,0), color=colours['6%'])
        hrc8 = Line2D((0,1),(0,0), color=colours['8%'])
        hrc10 = Line2D((0,1),(0,0), color=colours['10%'],linewidth=4)
        ax.legend([act1,mean1,hrc1,hrc2,hrc4,hrc6,hrc8,hrc10], ["Actual storage","Mean storage","1% HRC","2% HRC","4% HRC","6% HRC","8% HRC","10% HRC"])


    hrc_plotter(hrc,actual,means,ax1,'GWh',colours)
```

```python
history_repeats = ea.calc_mean_percentile(storage.sum(axis=1),80,future=True)
```

```python
hrc_plot(SI_HRC2['12/'+str(datetime.now().year-1):'2/1/'+str(datetime.now().year+1)],\
         history_repeats['12/'+str(datetime.now().year-1):'2/1/'+str(datetime.now().year+1)].Actual,\
         history_repeats['12/'+str(datetime.now().year-1):'2/1/'+str(datetime.now().year+1)]['mean'])
```
